```python
##### import current limit data #####

gazz = pd.read_csv("/data/Chen-Wang/gazz_import.csv")
gazz_gi_mass = gazz.gi_mass.to_numpy()
gazz_gi_gazz = gazz.gi_gazz.to_numpy()
gazz_photons_mass = gazz.photons_mass.to_numpy()
gazz_photons_gazz = gazz.photons_gazz.to_numpy()
gazz_NonresonantggF_mass = gazz.NonresonantggF_mass.to_numpy()
gazz_NonresonantggF_gazz = gazz.NonresonantggF_gazz.to_numpy()

gazy = pd.read_csv("/data/Chen-Wang/gazy_import.csv")
gazy_LEPZwidth_mass = gazy.LEPZwidth_mass.to_numpy()
gazy_LEPZwidth_gazy = gazy.LEPZwidth_gazy.to_numpy()
gazy_LEPZahad_mass = gazy.LEPZahad_mass.to_numpy()
gazy_LEPZahad_gazy = gazy.LEPZahad_gazy.to_numpy()
gazy_LHCtriboson_mass = gazy.LHCtriboson_mass.to_numpy()
gazy_LHCtriboson_gazy = gazy.LHCtriboson_gazy.to_numpy()
gazy_photons_mass = gazy.photons_mass.to_numpy()
gazy_photons_gazy = gazy.photons_gazy.to_numpy()
gazy_NonresonantggF_mass = gazy.NonresonantggF_mass.to_numpy()
gazy_NonresonantggF_gazy = gazy.NonresonantggF_gazy.to_numpy()
gazy_gi_mass = gazy.gi_mass.to_numpy()
gazy_gi_gazy = gazy.gi_gazy.to_numpy()
```

```python
##### plot our gazz result with current limit #####

y_upperlimit = 1
y_lowerlimit = 5e-6
x_upperlimit = 2000
x_lowerlimit = 1

plt.figure(figsize=(10,8))
plt.plot(mass_large, g_azz_300_large, label='Our work, Luminosity=300$(fb^{-1})$', color='blue')
plt.plot(mass_small, g_azz_300_small, color='blue')
plt.plot(mass_large, g_azz_3k_large, label='Our work, Luminosity=3000$(fb^{-1})$', color='orange')
plt.plot(mass_small, g_azz_3k_small, color='orange')

#plt.plot(gazz_gi_mass, gazz_gi_gazz, label="g.i.")
plt.plot(gazz_photons_mass, gazz_photons_gazz)
plt.fill_between(gazz_photons_mass, gazz_photons_gazz, y_upperlimit, alpha=0.2, label="photons (various)")
plt.plot(gazz_NonresonantggF_mass, gazz_NonresonantggF_gazz)
plt.fill_between(gazz_NonresonantggF_mass, gazz_NonresonantggF_gazz, y_upperlimit, alpha=0.2, label="Nonresonant ggF (CMS)")

plt.ylim(y_lowerlimit,y_upperlimit)
plt.xlim(x_lowerlimit,x_upperlimit)
plt.xscale("log")
plt.yscale("log")
plt.xlabel("ALP mass (GeV)")
plt.ylabel("$\\rmg_{aZZ} (GeV^{-1})$")
plt.legend(loc='best')
plt.show()
```
