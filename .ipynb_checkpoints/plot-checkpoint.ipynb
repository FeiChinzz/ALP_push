```python
##### plot the limit of parameters in different mass #####
##### note that mass=10 GeV has no background event left #####

mass = [100, 80, 50, 10, 1]
f_a_300 = [56316.4970567684, 64346.53837988309, 91422.25761046301, 127795.27360626447, 13730.589771704113]
g_azz_300 = [8.685909556961876e-05, 7.601962938738722e-05, 5.350556995477402e-05, 3.827684594244819e-05, 0.0003562556365991335]
g_azy_300 = [5.9121826456271114e-05, 5.174379616132217e-05, 3.64192949577118e-05, 2.605365653720195e-05, 0.00024249025140033248]

f_a_3k = [103041.72348327187, 118647.77379385554, 175492.17667036498, 404124.1388001958, 24440.941707505073]
g_azz_3k = [4.7472032053055857e-05, 4.122791219411252e-05, 2.7873607204656864e-05, 1.210420148255106e-05, 0.00020013958784975697]
g_azy_3k = [3.2312485205625385e-05, 2.8062339976140305e-05, 1.8972550393909476e-05, 8.238889603329359e-06, 0.00013622773645390943]

plt.scatter(mass, f_a_300, label='Luminosity=300$(fb^{-1})$')
plt.scatter(mass, f_a_3k, label='Luminosity=3000$(fb^{-1})$')
plt.xlabel("ALP mass (GeV)")
plt.ylabel("upper limit of $f_a ($GeV$)$")
plt.yscale("log")
plt.legend(loc='best')
plt.show()

plt.scatter(mass, g_azz_300, label='Luminosity=300$(fb^{-1})$')
plt.scatter(mass, g_azz_3k, label='Luminosity=3000$(fb^{-1})$')
plt.xlabel("ALP mass (GeV)")
plt.ylabel("lower limit of $g_{aZZ}$ ($GeV^{-1}$)")
plt.yscale("log")
plt.legend(loc='best')
plt.show()

plt.scatter(mass, g_azy_300, label='Luminosity=300$(fb^{-1})$')
plt.scatter(mass, g_azy_3k, label='Luminosity=3000$(fb^{-1})$')
plt.xlabel("ALP mass (GeV)")
plt.ylabel("lower limit of $g_{aZ\gamma}$ ($GeV^{-1}$)")
plt.yscale("log")
plt.legend(loc='best')
plt.show()
```

```python
N_sim_sig = 10000
N_sim_bg = 100000
Nselected_sig = 1626
Nselected_bg = 43
Xsection_sig = 506.49175    ### unit:fb
Xsection_bg = 99.09201    ### unit:fb
Luminosity = 3000    ### unit:fb^-1

print("In Luminosity = 3000(fb^-1)")

s = Xsection_sig * (Nselected_sig/N_sim_sig) * Luminosity
b = Xsection_bg * (Nselected_bg/N_sim_bg) * Luminosity

def significance(s,b):
    return np.sqrt(2*((s+b)*np.log(1+s/b)-s))

bot = 0
top = b
Z = 999
Zdiff = 0.00000000001

while abs(Z-2)>Zdiff:
    if significance((bot+top)/2, b)>2:
        top = (bot+top)/2
        Z = significance(top, b)
    elif significance((bot+top)/2, b)<2:
        bot = (bot+top)/2
        Z = significance(bot, b)
    #print(bot, top)
exp_s = (top+bot)/2
print(exp_s, "\n", significance(exp_s, b))
```
